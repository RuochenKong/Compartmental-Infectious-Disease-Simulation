```python
import numpy as np
import pandas as pd
```

```python
output_file = 'src_data/population.csv'
fout = open(output_file, 'w')
skip = 0
fout.write('GeoId,BlockGroup,CensusTract,County,State,Population\n')
with open('src_data/ACSDT5Y2019.B01001_2024-08-23T101550/ACSDT5Y2019.B01001-Data.csv', 'r') as f:
    for line in f.readlines():
        if skip < 2:
            skip += 1
            continue
        line = line.split('",')
        geoInfo = line[1][1:].split(', ')
        geoInfo_new = []
        for i,geo in enumerate(geoInfo):
            geo = geo.split(' ')
            if i == 2:
                geoInfo_new.append(geo[0])
            else:
                geoInfo_new.append(geo[-1])
        geoInfo_new = ','.join(geoInfo_new)
        fout.write('%s,%s,%s\n' % (line[0][1:].replace('1500000US',''), geoInfo_new, line[2][1:]))
fout.close()
```
