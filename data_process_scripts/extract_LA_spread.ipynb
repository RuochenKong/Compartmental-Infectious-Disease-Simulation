```python
import pandas as pd
import os
try: os.mkdir('spread_probability/LA')
except: pass
```

```python
df_list = []
for i in range(len(os.listdir('spread_probability'))):
    tmp_df = pd.read_csv('spread_probability/%d.csv'%(i+1))
    tmp_df['from_LA'] = tmp_df['from'].apply(lambda x: 1 if ('%012d'%x)[:2] == '22' else 0)
    tmp_df['to_LA'] = tmp_df['to'].apply(lambda x: 1 if ('%012d'%x)[:2] == '22' else 0)
    tmp_df['select'] = tmp_df['from_LA'] * tmp_df['to_LA']
    tmp_df['select'] = tmp_df['select'].apply(lambda x: x == 1)
    tmp_df = tmp_df[tmp_df['select']]
    print(f'File {i+1} done')
    if len(tmp_df) == 0: continue
    df_list.append(tmp_df.reset_index(drop=True))
    
```

```python
df_list
```

```python
df_la = pd.concat(df_list)
df_la
```

```python
def cal_compress(from_df, geoid):
    from_df = from_df.groupby('to')['prob'].sum().reset_index()
    from_df['from'] = len(from_df) * [geoid]
    from_df['prob'] = from_df['prob']/from_df['prob'].sum()
    from_df = from_df[['from','to','prob']]
    return from_df
```

```python
df_la['from'] = df_la['from'].apply(lambda x: ('%012d'%x)[:5])
df_la['to'] = df_la['to'].apply(lambda x: ('%012d'%x)[:5])
from_county = df_la['from'].unique().tolist()
grouped_df_la = df_la.groupby('from')

res = []
for county in from_county:
    res.append(cal_compress(grouped_df_la.get_group(county), county))
    
res_df = pd.concat(res)
res_df
```

```python
res_df.to_csv('spread_probability/LA/la_county.csv', index = False)
```
