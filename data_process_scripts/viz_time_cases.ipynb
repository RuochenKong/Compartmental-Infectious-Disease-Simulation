```python
import matplotlib.pyplot as plt
```

```python
def plot_res(init_cbg, init_cases, ax):
    cases = []
    times = []
    
    for line in open('res_%d_%d'%(init_cbg,init_cases)).readlines():
        line = line.strip().split(' ')
        if len(line) < 2:
            break
        cases.append(int(line[0]))
        times.append(float(line[1]))
    ax.scatter(cases, times)
```

```python
fig,ax = plt.subplots()
plot_res(1,10,ax)
plot_res(2,10,ax)
plot_res(4,10,ax)
plt.xlabel('Cases')
plt.ylabel('Time (s)')
plt.show()
```
