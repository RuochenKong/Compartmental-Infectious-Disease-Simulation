```python
import pandas as pd
import numpy as np
```

```python
# Age & Gender
selected_columns = ['B01001_001E','B01001_002E']+['B01001_0%02dE'%i for i in range(16,27)] + ['B01001_0%02dE'%i for i in range(40,50)]
df = pd.read_csv('ACSDT5Y2019.B01001_2024-08-23T101550/ACSDT5Y2019.B01001-Data.csv')
df['GeoId'] = df['GEO_ID'].apply(lambda x: x[9:]).to_list()
df = df[['GeoId'] + selected_columns].iloc[1:].reset_index(drop = True)
df.iloc[:,1:] = df.iloc[:,1:].apply(pd.to_numeric)
```

```python
df['total'] = df['B01001_001E']
df['male'] = df['B01001_002E']
df['female'] = df['B01001_026E']
df['old_male'] = df[['B01001_0%02dE'%i for i in range(16,26)]].sum(axis=1)
df['old_female'] = df[['B01001_0%02dE'%i for i in range(40,50)]].sum(axis=1)
df['young_male'] = df['male'] - df['old_male']
df['young_female'] = df['female'] - df['old_female']
df = df[['GeoId','total', 'young_male', 'young_female', 'old_male', 'old_female']].reset_index(drop = True)
```

```python
df
```

```python
df.iloc[df[df['total'] == 0].index,2:] = 1
df.iloc[df[df['total'] == 0].index,1] = 4
df
```

```python
df ['old_male'] = df['old_male']/df['total']
df ['old_female'] = df['old_female']/df['total']
df ['young_male'] = df['young_male']/df['total']
df ['young_female'] = df['young_female']/df['total']
```

```python
df = df[['GeoId', 'young_male', 'young_female', 'old_male', 'old_female']].reset_index(drop = True)
df.to_csv('age_gender.csv', index=False)
```

```python
# Race
selected_columns = ['B02001_001E', 'B02001_002E']
df = pd.read_csv('ACSDT5Y2023.B02001_2025-03-13T134907/ACSDT5Y2023.B02001-Data.csv')
df['GeoId'] = df['GEO_ID'].apply(lambda x: x[9:]).to_list()
df = df[['GeoId'] + selected_columns].iloc[1:].reset_index(drop = True)
df.iloc[:,1:] = df.iloc[:,1:].apply(pd.to_numeric)
df
```

```python
df['total'] = df['B02001_001E']
df['white'] = df['B02001_002E']
df['non-white'] = df['total']-df['white']
df = df[['GeoId', 'total', 'white', 'non-white']].reset_index(drop = True)
df.iloc[df[df['total'] == 0].index,2:] = 1
df.iloc[df[df['total'] == 0].index,1] = 2
df['white'] = df['white']/df['total']
df['non-white'] = df['non-white']/df['total']
df = df[['GeoId', 'white', 'non-white']].reset_index(drop = True)
df.to_csv('race.csv',index = False)
```

```python
# Income
selected_columns= ['S2001_C01_001E', 'S2001_C01_010E', 'S2001_C01_011E', 'S2001_C01_012E']
df = pd.read_csv('ACSST5Y2023.S2001_2025-03-13T135554/ACSST5Y2023.S2001-Data.csv')
df['GeoId'] = df['GEO_ID'].apply(lambda x: x[9:]).to_list()
df = df[['GeoId'] + selected_columns].iloc[1:].reset_index(drop = True)
df.iloc[:,1:] = df.iloc[:,1:].apply(pd.to_numeric)
df
```

```python
df['total'] = df['S2001_C01_001E']
df['rich'] = df['S2001_C01_010E'] + df['S2001_C01_011E'] + df['S2001_C01_012E']
df['non-rich'] = df['total']-df['rich']
df = df[['GeoId', 'total', 'rich', 'non-rich']].reset_index(drop = True)
df.iloc[df[df['total'] == 0].index,2:] = 1
df.iloc[df[df['total'] == 0].index,1] = 2
df['rich'] = df['rich']/df['total']
df['non-rich'] = df['non-rich']/df['total']
df = df[['GeoId', 'rich', 'non-rich']].reset_index(drop = True)
df.to_csv('income(census_tract).csv',index = False)
```

### Merger

```python
age_gender_df = pd.read_csv('age_gender.csv', dtype={'GeoId':str})
race_df = pd.read_csv('race.csv', dtype={'GeoId':str})
age_gender_race_df = age_gender_df.join(race_df.set_index('GeoId'), on = 'GeoId', how='outer')
age_gender_race_df
```

```python
age_gender_race_df.iloc[:,1:5].fillna(0.25)
age_gender_race_df.iloc[:,5:].fillna(0.5)
age_gender_race_df
```

```python
income_ct_df = pd.read_csv('income(census_tract).csv',  dtype={'GeoId':str})
income_ct_df
```

```python
income_df = pd.DataFrame()
income_df['GeoId'] = age_gender_race_df['GeoId']
income_df['rich'] = 0.5
income_df['non-rich'] = 0.5
income_df['GeoId_tract']  = income_df['GeoId'].apply(lambda x: x[:-1])
grouped_income_df = income_df.groupby('GeoId_tract')

for i in range(len(income_ct_df)):
    if i % 1000 == 0:  print(i)
    ct_id = income_ct_df.iloc[i,0]
    g_income_df = grouped_income_df.get_group(ct_id)
    income_df.loc[g_income_df.index, 'rich'] = income_ct_df.iloc[i,1]
    income_df.loc[g_income_df.index, 'non-rich'] = income_ct_df.iloc[i,2]
    
```

```python
income_df = income_df[['GeoId', 'rich', 'non-rich']].reset_index(drop = True)
income_df
```

```python
age_gender_race_income_df = age_gender_race_df.join(income_df.set_index('GeoId'), on = 'GeoId', how='outer')
age_gender_race_income_df
```

```python
age_gender_race_income_df[1:5] = age_gender_race_income_df[1:5].fillna(0.25)
age_gender_race_income_df[5:] = age_gender_race_income_df[5:].fillna(0.5)
age_gender_race_income_df
```
